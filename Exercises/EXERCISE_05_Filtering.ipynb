# Filtering Data Exercise
Answer the following questions using the `bestsellers.csv` dataset

```python
#Find the books written by Pete Souza

import pandas as pd
```

```python
Bestsellers = pd.read_csv("bestsellers.csv")
```

```python
Bestsellers
```

```python
Bestsellers[Bestsellers["Author"] == "Pete Souza"]
```

```python
# Find the books that are under 10 dollars

Cheap = Bestsellers["Price"] < 10
```

```python
Bestsellers[Cheap]
```

```python
#Find the books that have a price between 50 and 60 dollars

Budget = Bestsellers["Price"] >= 50
```

```python
Expensive = Bestsellers["Price"] <= 60
```

```python
Bestsellers[Budget & Expensive]
```

```python
#Find all the books written by Kristin Hannah, Andy Weir, or Delia Owens

Krist = Bestsellers["Author"] == "Kristin Hannah"
```

```python
Andy = Bestsellers["Author"] == "Andy Weir"
```

```python
Delia = Bestsellers["Author"] == "Delia Owens"
```

```python
Bestsellers[Krist | Andy | Delia]
```

```python
#Find the Non Fiction books that are rated 4.9

fic = Bestsellers["Genre"] == "Fiction"
```

```python
Good_rating = Bestsellers["User Rating"] == 4.9
```

```python
Bestsellers[~fic & Good_rating]
# using ~ for bitwise negation
```

```python
#Find the fiction book with the lowest User Rating

Bestsellers[fic].sort_values("User Rating").head(1)
```

```python
#Find 2012's top 5 Fiction books with the most Reviews

Books_2012 = Bestsellers["Year"] == 2012
```

```python
Bestsellers[Books_2012 & fic].sort_values("Reviews", ascending=False).head()
```

```python
#Create a bar plot showing the 5 authors who have the most books with a rating under 4.5

Lower_rating = Bestsellers["User Rating"] < 4.5
```

```python
Bestsellers[Lower_rating].value_counts("Author").head().plot(kind="bar")
```
